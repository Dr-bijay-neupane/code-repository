#Matrix Operation

```python
A = np.array([[2, -1],
              [0,  3]])

B = np.array([[ 1,  2],
              [-3,  4]])
```

```python
#additon of array A and B i.e A+B
c = A + B
print(c)
```

```python
#multication of A and B i.e A*B
AB = A @ B
print(AB)
```

```python
#3A−2B
result = 3*A - 2*B
result
```

#DETERMINANT



```python
import numpy.linalg as LA
```

```python
det_A = LA.det(A)
print(det_A)
```

```python
#since determinant is 6 meaning its inverse exist so
A_inv = LA.inv(A)
print(A_inv)
```

#Eigenvalues and Eigenvectors

```python
C = np.array([[ 4,  1],
              [-2,  1]])

eigenvalues, eigenvectors = LA.eig(C)


lambda1 = eigenvalues[0]
lambda2 = eigenvalues[1]

print(lambda1, lambda2)

print("eigen vectors for ",lambda1)
v1 = eigenvectors[:, 0]
print(v1)

v2= eigenvectors[ :, 1]
print("eigen vectors for ",lambda2)
print(v2)

```

and price: 30

Then, find the price of a car 3 with features
x1: 8
x2:  3.0
x3: 200
price:?

Hints:
1. Represent the given dataset as a matrix equation X.T*β=y, where X is the feature matrix, β is the coefficient vector, and y is the price vector  
2. Compute β (the coefficients) using the normal equation:
β=(X.T * X)^−1  X.T y

where,
X.T --> transpose matrix

```python
X = np.array([[4, 1.5, 100],
              [6, 2.0, 150]])

y = np.array([[20],
              [30]])
```

```python
XT_X = X.T @ X
print(XT_X)
LA.det(XT_X)
```


- X has shape (2 × 3)
- (XᵀX) is not invertible
- Price of car 3 cannot be uniquely predicted


```python
#pinv is pseudo-inverse which is done when a normal matrix inverse does not exist
beta_pinv = LA.pinv(X) @ y
beta_pinv

x3 = np.array([[8, 3.0, 200]])
price_car3 = x3 @ beta_pinv
price_car3

```
